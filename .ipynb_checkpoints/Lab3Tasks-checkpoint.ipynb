```python
# Import the library
from PIL import Image
import numpy as np
```

```python
# ── Load image ────────────────────────────────────
img = Image.open("cameraman.jpg")
display(img)
```

### 1. Scale (increase size)
Double the image dimensions using `Image.resize()` with high-quality resampling `(LANCZOS)`.

```python
# Get the size of the image
w, h = img.size
# scaling factors
cx, cy = 2, 2
```

```python
# ── 1. Increase size (scale ×2) ────────────────────
new_w= int(w*cx)
new_h= int(w*cy)
# Resize the image using PIL's built-in method
img_scaled = img.resize((new_w, new_h), Image.Resampling.LANCZOS)
```

```python

# Save the scaled image and print the sizes (The new image name should be "task1_1_scaled.jpg")
img_scaled.save("task1_1_scaled.jpg")
```

### 2. Rotate 120°
Rotate the image by 120 degrees, expanding the canvas to fit the full rotated image.

### 3. Shear

```python
# -- c. Get the image dimensions ────────────────────
w, h = img.size
```

```python
# -- d. define the shear matrix ──────────────────────────

# Choose X-axis or Y-axis shear. The shear factor controls how much the image slants — start with 0.5 then experiment.
shx= 0.5
shy= 0.5
```

```python
# -- e. Apply the shear transformation to the image ──────────────────────────
# PIL's transform() takes the inverse affine matrix:
# [ 1    shx   tx ]
# [ shy  1     ty ]
sheared = img.transform(
    (w,h),
    Image.AFFINE,
    (1, shx, 0,
     shy, 1, 0),
    resample=Image.Resampling.BICUBIC
)
```

```python
# -- f. Save the sheared image(The new image name should be "task1_3_sheared.jpg") 
sheared.save("task1_3_sheared.jpg")
```

### Experiment and compare
Try these:

— Change shear factor from 0.5 to 0.1, 0.3, 0.8 and compare results

— Switch from X-axis to Y-axis shear matrix

— Update the canvas multiplier to match your new factor

— Try combining X and Y shear in one matrix

```python
#changing shear factor
for shx in [0.1, 0.3, 0.5, 0.8]:
    sheared = img.transform(
        (w, h), Image.AFFINE,
        (1, shx, 0,
         0,  1, 0),
        resample=Image.Resampling.BICUBIC
    )
    sheared.save(f"shear_x_{shx}.jpg")
    display(sheared)
    print(f"shx = {shx}")

#Switch to Y-axis shear
shy = 0.5
sheared_y = img.transform(
    (w, h), Image.AFFINE,
    (1,   0, 0,
     shy, 1, 0),
    resample=Image.Resampling.BICUBIC
)
sheared_y.save("shear_y_0.5.jpg")
display(sheared_y)

#Update canvas to fit the shear (no cropping)
shx = 0.5
new_w = int(w + shx * h)   # expand canvas width
sheared_full = img.transform(
    (new_w, h), Image.AFFINE,
    (1, shx, 0,
     0,  1, 0),
    resample=Image.Resampling.BICUBIC
)
sheared_full.save("shear_expanded_canvas.jpg")
display(sheared_full)

#Combined X and Y shear
shx, shy = 0.3, 0.3
sheared_xy = img.transform(
    (w, h), Image.AFFINE,
    (1,   shx, 0,
     shy, 1,   0),
    resample=Image.Resampling.BICUBIC
)
sheared_xy.save("shear_xy_combined.jpg")
display(sheared_xy)
```

# Intensity Transformations
Negative · Log · Power Law (Gamma)

```python

# ── 1. Negative ──────────────────────────────────── 
# Method 1: NumPy array manipulation
arr=np.array(img)
negative_arr=255-arr
negative_img=Image.fromarray(negative_arr)

negative_img.save("task1_4_negative.jpg")
```

```python
# Method 2: PIL's ImageOps
from PIL import ImageOps
img_negative_ops= ImageOps.invert(img)
```

```python
# ── 2. Log transformation ──────────────────────────

# s = c · log(1 + r) --> We need to find c.
# We want the maximum output value (s) to be 255 when the maximum input value (r) is 255:
# 255 = c · log(1 + 255)
# c = 255 / log(1 + 255)
arr_float = arr.astype(np.float64)
c = 255 / np.log(1 + arr_float.max())


# Apply the log transformation to each pixel
log_transformed_arr= c*np.log(1+arr_float)
log_transformed_arr= Image.fromarray(np.uint8(log_transformed_arr))

log_transformed_arr.save("task1_5_log_transformed.jpg")
```

```python

# ── 3. Power-law / Gamma correction ───────────────
arr_float = arr.astype(np.float64) / 255.0  # normalize to [0, 1]

for gamma in [0.5, 1.0, 2.0]:
    gamma_corrected = np.power(arr_float, gamma)          # apply s = r^gamma
    gamma_corrected = np.uint8(gamma_corrected * 255)     # scale back to [0, 255]
    gamma_img = Image.fromarray(gamma_corrected)
    gamma_img.save(f"task1_5_gamma_{gamma}.jpg")
    print(f"Saved gamma={gamma}")
```
